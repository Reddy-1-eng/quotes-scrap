```python
###importing library

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline

```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
train_df.shape
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
test_df.shape
```

```python
final_df= pd.concat([train_df,test_df],ignore_index=True)
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
### feature engineering on date_of_the_journey feature 
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
### convertinf the data type 

final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Date'].astype(int)
final_df['Year']=final_df['Date'].astype(int)
```

```python
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis = 1 ,inplace=True)
final_df.head()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df.head()
```

```python
### creating a derived feature 'Arrival_hour' and ' Arrival_min' from 'Arrival_time feature

final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_minute']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minute']=final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df.drop('Dep_Time', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_minute']=final_df['Dep_minute'].astype(int)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
### feature engineering onm route feature 

final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'nan':1,'4 stops':4})
```

```python
final_df.head()
```

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head()
```

```python
final_df.head()
```

```python
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_minute']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.head()
```

```python
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(13343,axis=0,inplace=True)
```
