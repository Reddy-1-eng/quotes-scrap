**Creating Subplots**

it means that within one canvas i can create multi subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'bo')
plt.subplot(2,2,3)
plt.plot(x,y,'go--')
plt.subplot(2,2,4)
plt.plot(x,y,'b*--')
```

This code uses plt.subplot() to create a 2x2 grid of subplots, placing a different plot in each of the four cells. Here’s a breakdown:

plt.subplot(2, 2, 1):

Creates the first subplot in a 2x2 grid (i.e., 2 rows and 2 columns) and places it in the first position.

plt.plot(x, y, 'r--'): Plots y = x^2 with a red dashed line ('r--').

plt.subplot(2, 2, 2):

Creates the second subplot in the second position of the 2x2 grid.

plt.plot(x, y, 'bo'): Plots y = x^2 with blue circular
markers ('bo'), without connecting lines.
plt.subplot(2, 2, 3):

Creates the third subplot in the third position of the grid.
plt.plot(x, y, 'go--'): Plots y = x^2 with green circular markers connected by a dashed line ('go--').
plt.subplot(2, 2, 4):

Creates the fourth subplot in the last position of the grid.
plt.plot(x, y, 'b*--'): Plots y = x^2 with blue star markers connected by a dashed line ('b*--').

```python
## Compute the x and y coordinates for points on a sine curve

x=np.arange(0,4*np.pi,0.1)  ## taking the values from 0 to 4pi and the third parameter is the step size
y=np.sin(x)
plt.title('Sine wave form')

## plot the points using matplotlib

plt.plot(x,y)
plt.show()
```

```python
## Subplot

## Compute the x and y coordinates for points on a sine curve and cose curve

x=np.arange(0,5*np.pi,0.1)  ## taking the values from 0 to 4pi and the third parameter is the step size
y_sin=np.sin(x)
y_cos=np.cos(x)

## Set up the subplot grid that has height 2 and width 1
## and set the first subplot as active

plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine wave')

## Set the second subplot as active and make the second plot

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cos Wave')


## Show the figure

plt.show()
```

```python
#Bar Plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

This code creates a bar plot with two sets of bars (x, y and x2, y2). Here's how it works:

x = [2, 8, 10] and y = [11, 16, 9]:

These lists represent the positions (x) and heights (y) of the first set of bars.

x2 = [3, 9, 11] and y2 = [6, 15, 7]:

These lists represent the positions (x2) and heights (y2) of the second set of bars.

plt.bar(x, y, align='center'):

Plots a bar at each position in x with the corresponding height in y.

The align='center' parameter centers the bars on the x-coordinates.

plt.bar(x2, y2, color='g'):

Plots the second set of bars with x2 as positions and y2 as heights.

color='g' makes these bars green.
plt.title('Bar Graph'):

Sets the plot title to "Bar Graph."
plt.ylabel('Y axis') and plt.xlabel('X axis'):

Labels the y-axis and x-axis as "Y axis" and "X axis," respectively.
plt.show():

Displays the plot.

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

## Rectangular Box Plot

plt.boxplot(data,vert=True,patch_artist=True)
```

data = [np.random.normal(0, std, 100) for std in range(1, 4)]:

This list comprehension creates three datasets (one for each standard deviation in the range 1 to 3).

Each dataset contains 100 random numbers sampled from a normal (Gaussian) distribution with a mean of 0 and a standard deviation of std.

plt.boxplot(data, vert=True, patch_artist=True):

plt.boxplot(data): Creates a box plot for each dataset in data.

vert=True: Ensures that the boxes are vertical.

patch_artist=True: Fills each box with color for better visual distinction.

The resulting box plot displays the spread (interquartile range), median, and potential outliers for each of the three datasets, showing how the distributions vary with increasing standard deviation.

```python
#PIE CHART
## DATA TO PLOT

labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #explode 1st list

#Plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.2f%%',shadow=True)

plt.axis('equal')
plt.show()
```

Data and Settings:

labels = ['Python', 'C++', 'Ruby', 'Java']: Names of the categories displayed in the pie chart.

sizes = [215, 130, 245, 210]: Values representing the relative size or proportion of each category.

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']: Custom colors for each slice of the pie chart.

explode = (0.1, 0, 0, 0): Specifies the "exploding" effect for the first slice (Python), offsetting it slightly from the center.

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True):

plt.pie(): Creates the pie chart with the specified sizes, labels, colors, and explode effect.

autopct='%1.1f%%': Adds percentages to each slice with one decimal place.

shadow=True: Adds a shadow effect for visual depth.
plt.axis('equal'):

Sets the aspect ratio to make the pie chart circular rather than oval-shaped.

plt.show():

Displays the pie chart.
This pie chart will display each category with its corresponding color, label, and percentage, with the "Python" slice separated slightly for emphasis.

**Correlation with Heatmap**

A correlation heatmap uses colored cells , typically in a monochromatic scale,to show a 2D corrlation matrix(table) between two discrete dimensions or event types. It is very important in feature scaling

Whenever we are trying to apply correlation the value will be ranging between -1 and 1 and that is called Spearson's Coefficient Correlation


Correlation is a statistical term that describes the relationship between two variables. In simple words, it tells us how closely related two things are.

Positive Correlation: When one variable increases, the other variable tends to increase as well (e.g., height and weight).

Negative Correlation: When one variable increases, the other variable tends to decrease (e.g., the number of hours spent watching TV and grades in school).

No Correlation: There is no predictable relationship between the two variables (e.g., the color of a car and its fuel efficiency).

In summary, correlation helps us understand if and how two things are connected.

```python
numeric_df = df.select_dtypes(include=[np.number])  # Only keep numeric columns
```

```python
numeric_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))  # Optional: Set the size of the plot
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')  # `annot=True` adds the correlation coefficients to the heatmap
plt.show()
```
