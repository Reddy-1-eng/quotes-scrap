```python
def water_jug_solver(capacity1, capacity2, target1):
    
    jug1, jug2 = 0, 0
    steps = []

    
    def print_state():
        steps.append((jug1, jug2))
    
    
    def is_goal():
        return jug1 == target1
    

    
    while not is_goal():
        if jug1 == 0:
            
            jug1 = capacity1
            print_state()
        elif jug2 == capacity2:
            
            jug2 = 0
            print_state()
        else:
            
            pour_amount = min(jug1, capacity2 - jug2)
            jug1 -= pour_amount
            jug2 += pour_amount
            print_state()
        
        
        if is_goal():
            break
        
        if jug1 == 0:
           
            jug1 = capacity1
            print_state()
        
        if jug2 == capacity2:
           
            jug2 = 0
            print_state()
        
       
        pour_amount = min(jug2, capacity1 - jug1)
        jug2 -= pour_amount
        jug1 += pour_amount
        print_state()

    return steps

def print_solution(steps):
    if not steps:
        print("No solution exists.")
    else:
        print("Solution found:")
        for step in steps:
            print(f"Jug1: {step[0]}, Jug2: {step[1]}")

capacity1 = 4
capacity2 = 3
target1 = 2

steps = water_jug_solver(capacity1, capacity2, target1, )
print_solution(steps)
```

```python
import random

class SimpleVacuumCleaner:
    def __init__(self, grid_size):  # Corrected __init__ method
        self.grid_size = grid_size
        self.grid = [[random.choice(['Clean', 'Dirty']) for _ in range(grid_size)] for _ in range(grid_size)]
        self.position = (0, 0)  
        self.cleaned_cells = 0

    def display_grid(self):
        for row in self.grid:
            print(' '.join(row))
        print()

    def clean(self):
        x, y = self.position
        if self.grid[x][y] == 'Dirty':
            self.grid[x][y] = 'Clean'
            self.cleaned_cells += 1
            print(f"Cleaned cell at position {self.position}.")

    def move(self):
        x, y = self.position
        if y < self.grid_size - 1:
            self.position = (x, y + 1)
        elif x < self.grid_size - 1:
            self.position = (x + 1, 0)
        print(f"Moved to position {self.position}.")

    def run(self, steps):
        for _ in range(steps):
            self.clean()
            self.move()
            self.display_grid()
        
        print(f"Total cleaned cells: {self.cleaned_cells}")


grid_size = 3  
vacuum_cleaner = SimpleVacuumCleaner(grid_size)
vacuum_cleaner.display_grid()
vacuum_cleaner.run(9)
```

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt

def best_first_search(graph, start, goal, heuristic):
    priority_queue = []
    heapq.heappush(priority_queue, (heuristic[start], start))
    explored = set()
    parent = {}
    parent[start] = None

    while priority_queue:
        _, current = heapq.heappop(priority_queue)
        
        if current == goal:
            path = []
            while current is not None:
                path.append(current)
                current = parent[current]
            path.reverse()
            return path
        
        explored.add(current)
        
        for neighbor in graph[current]:
            if neighbor not in explored:
                heapq.heappush(priority_queue, (heuristic[neighbor], neighbor))
                if neighbor not in parent:
                    parent[neighbor] = current
    
    return None

# Define the graph
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G'],
    'D': ['H'],
    'E': [],
    'F': [],
    'G': [],
    'H': []
}

# Define the heuristic values
heuristic = {
    'A': 6,
    'B': 4,
    'C': 5,
    'D': 3,
    'E': 4,
    'F': 2,
    'G': 6,
    'H': 0
}

start_node = 'A'
goal_node = 'H'

# Perform Best-first search
path = best_first_search(graph, start_node, goal_node, heuristic)
print("Path:", path)

# Visualize the graph and the path
def visualize_graph(graph, path):
    G = nx.Graph()
    
    # Add edges to the graph
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)
    
    # Define node positions for visualization
    pos = nx.spring_layout(G)
    
    # Draw the graph
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold')
    
    # Highlight the path
    if path:
        edge_list = list(zip(path, path[1:]))
        nx.draw_networkx_edges(G, pos, edgelist=edge_list, edge_color='r', width=2)
    
    # Show the plot
    plt.title("Best-first Search Path Visualization")
    plt.show()

# Call the visualization function
visualize_graph(graph, path)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the triangular membership function
def triangular(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)

# Define the range of heights
x = np.linspace(120, 220, 1000)

# Define the membership functions
very_short = triangular(x, 120, 130, 140)
short = triangular(x, 130, 150, 170)
normal = triangular(x, 150, 170, 190)
tall = triangular(x, 170, 190, 210)
very_tall = triangular(x, 190, 210, 220)

# Plotting the membership functions
plt.plot(x, very_short, label="Very Short")
plt.plot(x, short, label="Short")
plt.plot(x, normal, label="Normal")
plt.plot(x, tall, label="Tall")
plt.plot(x, very_tall, label="Very Tall")

# Add labels and legends
plt.title('Fuzzy Membership Functions for Height')
plt.xlabel('Height (cm)')
plt.ylabel('Membership Degree')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))

x = np.linspace(0, 10, 500)

white = gaussian(x, c=2, sigma=1)

moderate = gaussian(x, c=5, sigma=1)

black = gaussian(x, c=8, sigma=1)

plt.figure(figsize=(10, 6))
plt.plot(x, white, label="White", color='blue')
plt.plot(x, moderate, label="Moderate", color='green')
plt.plot(x, black, label="Black", color='black')

plt.title('Gaussian Membership Functions for Race of People')
plt.xlabel('Race Score')
plt.ylabel('Membership Degree')
plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# Define input variables
degree_of_dirt = ctrl.Antecedent(np.arange(0, 101, 1), 'degree_of_dirt')  # 0 to 100 scale
type_of_dirt = ctrl.Antecedent(np.arange(0, 101, 1), 'type_of_dirt')      # 0 to 100 scale

# Define output variable
wash_time = ctrl.Consequent(np.arange(0, 61, 1), 'wash_time')  # 0 to 60 minutes

# Membership functions for degree_of_dirt (Input)
degree_of_dirt['small'] = fuzz.trimf(degree_of_dirt.universe, [0, 0, 50])
degree_of_dirt['medium'] = fuzz.trimf(degree_of_dirt.universe, [25, 50, 75])
degree_of_dirt['large'] = fuzz.trimf(degree_of_dirt.universe, [50, 100, 100])

# Membership functions for type_of_dirt (Input)
type_of_dirt['not_greasy'] = fuzz.trimf(type_of_dirt.universe, [0, 0, 50])
type_of_dirt['medium'] = fuzz.trimf(type_of_dirt.universe, [25, 50, 75])
type_of_dirt['greasy'] = fuzz.trimf(type_of_dirt.universe, [50, 100, 100])

# Membership functions for wash_time (Output)
wash_time['very_short'] = fuzz.trimf(wash_time.universe, [0, 0, 15])
wash_time['short'] = fuzz.trimf(wash_time.universe, [10, 15, 30])
wash_time['medium'] = fuzz.trimf(wash_time.universe, [20, 30, 45])
wash_time['long'] = fuzz.trimf(wash_time.universe, [40, 45, 60])
wash_time['very_long'] = fuzz.trimf(wash_time.universe, [50, 60, 60])

# Define rules
rule1 = ctrl.Rule(degree_of_dirt['large'] & type_of_dirt['greasy'], wash_time['very_long'])
rule2 = ctrl.Rule(degree_of_dirt['medium'] & type_of_dirt['greasy'], wash_time['long'])
rule3 = ctrl.Rule(degree_of_dirt['small'] & type_of_dirt['greasy'], wash_time['long'])

rule4 = ctrl.Rule(degree_of_dirt['large'] & type_of_dirt['medium'], wash_time['long'])
rule5 = ctrl.Rule(degree_of_dirt['medium'] & type_of_dirt['medium'], wash_time['medium'])
rule6 = ctrl.Rule(degree_of_dirt['small'] & type_of_dirt['medium'], wash_time['medium'])

rule7 = ctrl.Rule(degree_of_dirt['large'] & type_of_dirt['not_greasy'], wash_time['medium'])
rule8 = ctrl.Rule(degree_of_dirt['medium'] & type_of_dirt['not_greasy'], wash_time['short'])
rule9 = ctrl.Rule(degree_of_dirt['small'] & type_of_dirt['not_greasy'], wash_time['very_short'])

# Control System Creation and Simulation
wash_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
wash_sim = ctrl.ControlSystemSimulation(wash_ctrl)

# Input values for degree of dirt and type of dirt
wash_sim.input['degree_of_dirt'] = 70  # Example input for degree of dirt
wash_sim.input['type_of_dirt'] = 50    # Example input for type of dirt (medium)

# Compute the output
wash_sim.compute()

# Output wash time
print(f"Recommended wash time: {wash_sim.output['wash_time']} minutes")

# Plot membership functions for degree_of_dirt
degree_of_dirt.view()

# Plot membership functions for type_of_dirt
type_of_dirt.view()

# Plot membership functions for wash_time
wash_time.view()

# Visualize the final result for wash_time
wash_time.view(sim=wash_sim)

# Show all plots
plt.show()
```

```python
def andnot_neural_model(X1,X2):
    w1 = 2
    w2 = -2
    bias = 1.5

    weighted_sum = X1 * w1 + X2 * w2
    if weighted_sum >= bias:
        return 1
    else:
        return 0

print("X1 | X2 | Y")
for X1 in [0,1]:
    for X2 in [0,1]:
        Y = andnot_neural_model(X1, X2)
        print(f"{X1} | {X2} | {Y}")
```
