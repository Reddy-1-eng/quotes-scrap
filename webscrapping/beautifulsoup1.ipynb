```python
import csv
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = "https://webscraper.io/test-sites/e-commerce/allinone"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
products = soup.find_all("div", class_="card-body")

with open("products.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Price", "Description", "Image URL", "Reviews"])
    
    for product in products:
        name = product.find("a", class_="title")
        name = name.text.strip() if name else "N/A"
        
        price = product.find("h4", class_="float-end price")
        price = price.text.strip() if price else "N/A"
        
        description = product.find("p", class_="description")
        description = description.text.strip() if description else "N/A"
        
        reviews = product.find("p", class_="float-end")
        reviews = reviews.text.strip().split()[0] if reviews else "N/A"
        
        image = product.find("img", class_="img-fluid card-img-top image img-responsive")
        image_url = image["src"] if image else "N/A"
        if image_url != "N/A" and image_url.startswith("/"):
            image_url = urljoin("https://webscraper.io", image_url)
        
        writer.writerow([name, price, description, image_url, reviews])
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd

url = "https://webscraper.io/test-sites/e-commerce/allinone"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
products = soup.find_all("div", class_="card-body")

data = []
for product in products:
    name = product.find("a", class_="title")
    name = name.text.strip() if name else "N/A"
    
    price = product.find("h4", class_="float-end price")
    price = price.text.strip() if price else "N/A"
    
    description = product.find("p", class_="description")
    description = description.text.strip() if description else "N/A"
    
    reviews = product.find("p", class_="float-end")
    reviews = reviews.text.strip().split()[0] if reviews else "N/A"
    
    image = product.find("img", class_="img-fluid card-img-top image img-responsive")
    image_url = image["src"] if image else "N/A"
    if image_url != "N/A" and image_url.startswith("/"):
        image_url = urljoin("https://webscraper.io", image_url)
    
    data.append([name, price, description, image_url, reviews])

df = pd.DataFrame(data, columns=["Name", "Price", "Description", "Image URL", "Reviews"])
```
